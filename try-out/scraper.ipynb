```python
import pandas
import requests
from bs4 import BeautifulSoup
```

```python
url_1 = "http://www.mkulimayoung.com/market"

# create request and get content
rq = requests.get(url_1)
c = rq.content
```

```python
c
```

```python
# beatify
soup = BeautifulSoup(c, "html.parser")
print(soup.prettify())
```

```python
# read all the 'div'
all = soup.find_all("div",{"class":"row"})
```
